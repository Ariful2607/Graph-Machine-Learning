```python
import pandas as pd

dataset = pd.read_csv('World Development Indicators.csv')
dataset.head()
```

```python
dataset = dataset[dataset['Country Name'] == 'Indonesia']
dataset
```

```python
dataset = dataset.drop(columns=['Country Name', 'Country Code', 'Indicator Code'])
dataset
```

```python
# Mengubah kolom tahun menjadi indeks
dataset = dataset.set_index('Indicator Name').T
dataset.head()
```

```python
dataset
```

```python
import torch
from torch_geometric.data import Data

node_features = torch.tensor(dataset['Cereal yield (kg per hectare)'].dropna().values, dtype=torch.float).view(-1, 1)

# Membuat edge connections (contoh: hubungan antar tahun secara berurutan)
num_nodes = node_features.size(0)
edge_index = torch.tensor([[i, i + 1] for i in range(num_nodes - 1)] + [[i + 1, i] for i in range(num_nodes - 1)], dtype=torch.long).t()

# Membuat label (jika ada, misalnya prediksi kategori tertentu)
labels = torch.zeros(num_nodes, dtype=torch.long)

# Membuat objek graf
data = Data(x=node_features, edge_index=edge_index, y=labels)

print(data)
```

```python
import torch
from torch_geometric.nn import GCNConv

import torch.nn.functional as F
import matplotlib.pyplot as plt

# Definisikan model GNN
class GNN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Inisialisasi model, optimizer, dan loss function
input_dim = node_features.size(1)
hidden_dim = 16
output_dim = 1
model = GNN(input_dim, hidden_dim, output_dim)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = torch.nn.MSELoss()

# Data untuk pelatihan
x = node_features
y = node_features  # Menggunakan data asli sebagai target
edge_index = edge_index

# Pelatihan model
model.train()
for epoch in range(200):
    optimizer.zero_grad()
    out = model(x, edge_index)
    loss = loss_fn(out, y)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')

# Prediksi
model.eval()
predictions = model(x, edge_index).detach().numpy()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Hitung RMSE
rmse = np.sqrt(mean_squared_error(y.numpy(), predictions))
print(f'RMSE: {rmse}')
```

```python
# Tampilkan hasil prediksi dalam grafik dengan time step berupa tahun
plt.figure(figsize=(10, 6))

# Gunakan indeks dataset (tahun) sebagai sumbu x
# Bersihkan indeks dataset dengan menghapus nilai yang tidak dapat dikonversi ke integer
dataset_cleaned = dataset[dataset.index.str.isnumeric()]
years = dataset_cleaned.index.astype(int)  # Pastikan indeks berupa integer (tahun)

# Ensure years matches the length of y and predictions
years = years[:len(y.numpy())]

plt.plot(years, y.numpy(), label='Actual', marker='o')
plt.plot(years, predictions, label='Predicted', marker='x')

plt.xlabel('Year')
plt.ylabel('Cereal Yield (kg per hectare)')
plt.title('GNN Prediction of Cereal Yield')
plt.legend()
plt.show()
```

```python

# Forecast future values
forecast_horizon = 10
historical_mean = node_features[:, 0].mean().item()
new_node_features = torch.full((forecast_horizon, 1), historical_mean, dtype=torch.float)

new_edges = [[i, i + 1] for i in range(num_nodes - 1, num_nodes + forecast_horizon - 1)]
new_edges += [[i + 1, i] for i in range(num_nodes - 1, num_nodes + forecast_horizon - 1)]
extended_edge_index = torch.cat([edge_index, torch.tensor(new_edges, dtype=torch.long).t()], dim=1)
extended_node_features = torch.cat([node_features, new_node_features], dim=0)

model.eval()
extended_predictions = model(extended_node_features, extended_edge_index).detach().numpy()

# Prepare forecast years
forecast_years = range(int(years[-1].item()) + 1, int(years[-1].item()) + 1 + forecast_horizon)
forecast_values = extended_predictions[-forecast_horizon:]

# Visualize results
plt.figure(figsize=(12, 6))
# Use years[:len(y)] to match the dimensions of y and predictions
plt.plot(years[:len(y)], y.numpy(), label='Actual', marker='o')
plt.plot(years[:len(predictions)], predictions, label='Predicted', marker='x')
plt.plot(forecast_years, forecast_values, label='Forecast', marker='*', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Cereal Yield (kg per hectare)')
plt.title('GNN Prediction and Forecast of Cereal Yield')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# --- Improved Forecast Initialization ---
def exponential_smoothing_initializer(historical_data, forecast_horizon):
    """Initialize future nodes using Holt-Winters exponential smoothing."""
    model = ExponentialSmoothing(historical_data.numpy(), trend='add', seasonal=None)
    fit = model.fit()
    return torch.tensor(fit.forecast(forecast_horizon), dtype=torch.float).view(-1, 1)

# Replace mean with smoothed forecasts
new_node_features = exponential_smoothing_initializer(node_features[:, 0], forecast_horizon)

# --- Time-Decaying Edge Weights (Optional) ---
def add_decaying_edges(num_nodes, forecast_horizon, decay_rate=0.9):
    """Edges lose influence exponentially over time."""
    new_edges = []
    edge_weights = []
    for i in range(num_nodes - 1, num_nodes + forecast_horizon - 1):
        new_edges.append([i, i + 1])
        new_edges.append([i + 1, i])
        edge_weights.extend([decay_rate**(i+1), decay_rate**(i+1)]) 
    return torch.tensor(new_edges).t(), torch.tensor(edge_weights)

extended_edge_index, edge_weight = add_decaying_edges(num_nodes, forecast_horizon)
extended_edge_index = torch.cat([edge_index.long(), extended_edge_index], dim=1)

# --- Monte Carlo Dropout for Uncertainty (PyTorch) ---
def mc_dropout_predict(model, x_test, edge_index): # Enable dropout at test time!
     model.train() 
     predictions = [model(x_test, edge_index).detach().numpy() for _ in range(100)] # Stochastic forward passes 
     return np.stack(predictions)

mc_predictions = mc_dropout_predict(model.eval(), extended_node_features, extended_edge_index)
mean_preds = mc_predictions.mean(axis=0)[-forecast_horizon:]
std_preds = mc_predictions.std(axis=0)[-forecast_horizon:]

# --- Visualization with Confidence Intervals ---
plt.figure(figsize=(12,6))
plt.plot(years[:len(y)], y.numpy(), label='Actual', marker='o')
plt.plot(years[:len(predictions)], predictions, label='Predicted', marker='x')
plt.plot(forecast_years[-10:], mean_preds, label='Forecast', marker='*', linestyle='--')
plt.fill_between(
     forecast_years[-10:],
     mean_preds.squeeze() - std_preds.squeeze(),
     mean_preds.squeeze() + std_preds.squeeze(),
     alpha=0.2,
 )
```
